```python
x = df[["Previous Scores","Hours Studied","Sleep Hours","Sample Question Papers Practiced"]]
```

```python
y = df["Performance Index"]
```

```python
x = np.array(x)
y = np.array(y)
```

```python
n = len(x)
split = int(0.8 * n)
```

```python
x_train = x[:split]
y_train = y[:split]
```

```python
x_test = x[split:]
y_test = y[split:]
```

```python
mean = np.mean(x_train,axis=0)
std = np.std(x_train,axis=0)
x_train_norm = (x_train-mean)/std
x_test_norm = (x_test-mean)/std
```

```python
def pred(w,b,x):
    y_pred = np.dot(x,w)+b
    return y_pred
```

```python
def cost(y_train,y_pred):
    err = y_pred - y_train
    cost = np.mean(np.square(err))
    return cost
```

```python
def gradient(err,x_train_norm):
    dw = 2*(np.mean(x_train_norm*err[:,np.newaxis],axis=0))
    db = 2*(np.mean(err))
    return dw,db
```

```python
def gradient_descent(w,b,dw,db,a):
    w = w - a * dw
    b = b - a * db
    return w,b
```

```python
def linear_regression(x_train_norm,y,w,b,a,I):
    for i in range(I):
        y_pred = pred(w,b,x_train_norm)
        Cost = cost(y,y_pred)
        err = y_pred - y
        dw,db = gradient(err,x_train_norm)
        w,b = gradient_descent(w,b,dw,db,a)
        if(i%100==0):
            print(f'iteration: {i}----------- cost: {Cost}')
    return w,b

    

```

```python
w = np.zeros(x_train_norm.shape[1])
b = 0
a = 0.001
I = 2000
n_w,n_b = linear_regression(x_train_norm,y_train,w,b,a,I)
```

```python
y_testpred = pred(n_w,n_b,x_test_norm)
```

```python
def r2score(y_test_norm,y_pred):
    num = np.sum(np.square(y_test-y_testpred))
    den = np.sum(np.square(y_test-np.mean(y_test)))
    r2 = 1-(num/den)
    return r2
```
